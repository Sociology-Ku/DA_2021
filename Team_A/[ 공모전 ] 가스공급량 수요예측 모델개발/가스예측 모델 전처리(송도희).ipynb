## 적설 결측치 채우기 <*계절에 따른 평균값> 

```python
# 계절에 따른 적설의 평균값.
weather.groupby("계절").mean("적설(cm)")
```

```python
# 인덱스 하나를 만듬.
# 여름에는 눈이 않오기때문에 결과 값이 nan이 나오는게 맞음. (나중에 따로 0으로 채워줄 예정)
weather.groupby("계절").mean()["적설(cm)"].reset_index()
```

```python
fill_mena_func = lambda g : g.fillna(g.mean())
weather.groupby("계절")["적설(cm)"].apply(fill_mena_func)
```

```python
weather[["계절" ,"적설(cm)"]].groupby("계절").apply(fill_mena_func)
```

```python
weather["적설(cm)"] = weather["적설(cm)"].fillna(weather.groupby("계절")["적설(cm)"].transform("mean"))
```

```python
weather
```

## 여름 결측치 0으로 채워주기

```python
# 다운
from tqdm import tqdm
```

```python
#tqdm을 쓰면 for문의 진행상황을 알수가 있다.

weather = weather.reset_index()

for i in tqdm(range(len(weather))):
  if weather['계절'][i] == '여름':
    weather['적설(cm)'][i] = 0
```
